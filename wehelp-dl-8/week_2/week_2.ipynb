```python
import math


class Point:
    def __init__(self, x: float, y: float):
        self.x = x
        self.y = y

    def distance_to(self, other: "Point") -> float:
        return math.hypot(
            self.x - other.x,
            self.y - other.y
        )


class Line:
    def __init__(self, start: Point, end: Point):
        if start.x == end.x and start.y == end.y:
            raise ValueError("A line must have two different points.")

        self.start = start
        self.end = end

    def is_parallel_to(self, other: "Line") -> bool:
        dx1 = self.end.x - self.start.x
        dy1 = self.end.y - self.start.y

        dx2 = other.end.x - other.start.x
        dy2 = other.end.y - other.start.y

        cross_product = dx1 * dy2 - dy1 * dx2

        return math.isclose(cross_product, 0.0, abs_tol=1e-9)

    def is_perpendicular_to(self, other: "Line") -> bool:
        dx1 = self.end.x - self.start.x
        dy1 = self.end.y - self.start.y

        dx2 = other.end.x - other.start.x
        dy2 = other.end.y - other.start.y

        dot_product = dx1 * dx2 + dy1 * dy2

        return math.isclose(dot_product, 0.0, abs_tol=1e-9)


class Circle:
    def __init__(self, center: Point, radius: float):
        if radius <= 0:
            raise ValueError("Radius must be greater than zero.")

        self.center = center
        self.radius = radius

    def area(self) -> float:
        return math.pi * self.radius ** 2

    def intersects(self, other: "Circle") -> bool:
        distance = self.center.distance_to(other.center)

        minimum_distance = abs(self.radius - other.radius)
        maximum_distance = self.radius + other.radius

        return minimum_distance <= distance <= maximum_distance


class Polygon:
    def __init__(self, vertices: list[Point]):
        if len(vertices) < 3:
            raise ValueError(
                "A polygon must contain at least three vertices."
            )

        self.vertices = vertices

    def perimeter(self) -> float:
        total = 0.0

        for i in range(len(self.vertices)):
            current_point = self.vertices[i]
            next_point = self.vertices[
                (i + 1) % len(self.vertices)
            ]

            total += current_point.distance_to(next_point)

        return total
```

```python
# Line A points
point_a1 = Point(-6, 1)
point_a2 = Point(2, 4)

# Line B points
point_b1 = Point(-6, -1)
point_b2 = Point(2, 2)

# Line C points
point_c1 = Point(-4, -4)
point_c2 = Point(-1, 6)

# Create lines
line_a = Line(point_a1, point_a2)
line_b = Line(point_b1, point_b2)
line_c = Line(point_c1, point_c2)

# Circle centers
circle_a_center = Point(6, 3)
circle_b_center = Point(8, 1)

# Create circles
circle_a = Circle(circle_a_center, 2)
circle_b = Circle(circle_b_center, 1)

# Polygon A vertices
polygon_point_1 = Point(-1, -2)
polygon_point_2 = Point(2, 0)
polygon_point_3 = Point(5, -1)
polygon_point_4 = Point(4, -4)

polygon_a = Polygon([
    polygon_point_1,
    polygon_point_2,
    polygon_point_3,
    polygon_point_4
])

print("Are Line A and Line B parallel?", line_a.is_parallel_to(line_b))
print("Are Line C and Line A perpendicular?", line_c.is_perpendicular_to(line_a))
print(f"Area of Circle A: {circle_a.area():.4f}")
print("Do Circle A and Circle B intersect?", circle_a.intersects(circle_b))
print(f"Perimeter of Polygon A: {polygon_a.perimeter():.4f}")
```

### Task 2

```python
import math


class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance_to(self, other):
        return math.hypot(
            self.x - other.x,
            self.y - other.y
        )

    def move_by(self, dx, dy):
        self.x += dx
        self.y += dy


class Enemy:
    DEFAULT_LIFE_POINTS = 10

    def __init__(self, label, position, move_x, move_y):
        self.label = label
        self.position = position
        self.move_x = move_x
        self.move_y = move_y
        self.life_points = self.DEFAULT_LIFE_POINTS

    @property
    def is_alive(self):
        return self.life_points > 0

    def move(self):
        if not self.is_alive:
            return

        self.position.move_by(
            self.move_x,
            self.move_y
        )

    def take_damage(self, damage):
        if not self.is_alive:
            return

        self.life_points -= damage

        if self.life_points < 0:
            self.life_points = 0


class Tower:
    def __init__(
        self,
        label,
        position,
        attack_points,
        attack_range
    ):
        self.label = label
        self.position = position
        self.attack_points = attack_points
        self.attack_range = attack_range

    def is_enemy_in_range(self, enemy):
        distance = self.position.distance_to(
            enemy.position
        )

        return distance <= self.attack_range

    def attack(self, enemies):
        for enemy in enemies:
            if (
                enemy.is_alive
                and self.is_enemy_in_range(enemy)
            ):
                enemy.take_damage(self.attack_points)


class BasicTower(Tower):
    def __init__(self, label, position):
        super().__init__(
            label=label,
            position=position,
            attack_points=1,
            attack_range=2
        )


class AdvancedTower(Tower):
    def __init__(self, label, position):
        super().__init__(
            label=label,
            position=position,
            attack_points=2,
            attack_range=4
        )
```

```python
enemy_1 = Enemy(label="E1", position=Point(-10, 2), move_x=2, move_y=-1)
enemy_2 = Enemy(label="E2", position=Point(-8, 0), move_x=3, move_y=1)
enemy_3 = Enemy(label="E3", position=Point(-9, -1), move_x=3, move_y=0)
enemies = [enemy_1, enemy_2, enemy_3]

tower_1 = BasicTower("T1", Point(-3, 2))
tower_2 = BasicTower("T2", Point(-1, -2))
tower_3 = BasicTower("T3", Point(4, 2))
tower_4 = BasicTower("T4", Point(7, 0))

advanced_tower_1 = AdvancedTower("A1", Point(1, 1))
advanced_tower_2 = AdvancedTower("A2", Point(4, -3))

towers = [tower_1, tower_2, tower_3, tower_4, advanced_tower_1, advanced_tower_2]

TOTAL_TURNS = 10

for turn in range(1, TOTAL_TURNS + 1):
    for enemy in enemies:
        enemy.move()

    for tower in towers:
        tower.attack(enemies)

for enemy in enemies:
    print(
        f"{enemy.label}: "
        f"final position = "
        f"({enemy.position.x}, {enemy.position.y}), "
        f"life points = {enemy.life_points}"
    )
```
